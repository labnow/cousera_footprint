# 2. OpenCV

### Import Libraries

### Import OpenCV

```python
import cv2
```

##### Get the image with the <b>imread</b>

```python
img = cv2.imread('images/rock.jpg')
```

### Always check the type

```python
type(img)
```

##### Get the image <b>shape</b>

```python
img.shape
```

##### Let's see the image with the <b>imshow</b>

```python
plt.imshow(img)
```

##### Until now we were working with <U>Matplotlib</u> and <b>RGB</b>. <br>
##### <u>OpenCV</u> is reading the channel as <b>BGR</b>. <br>
##### We will <b>convert</b> OpenCV to the channels of the photo.

```python
img_fix = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img_fix)
```

##### Scale it to <b>Gray</b> and check the <b>Shape</b>

```python
img_grey = cv2.imread('images/rock.jpg', cv2.IMREAD_GRAYSCALE)
plt.imshow(img_grey, cmap='gray')
# cv2.imshow('dsa',img_grey)
```

```python
img_grey.shape
```

### Resize the image

##### To resize the image we change the Height Width --> Width Height

```python
img_new = cv2.resize(img_fix, (400,800))
plt.imshow(img_new)
```

### Resize with Ratio

```python
resize_ratio = 0.5
img_fix.shape
```

```python
img2 = cv2.resize(img_fix, (0,0), img_fix, resize_ratio, resize_ratio)
img2.shape, img_fix.shape
```

```python
plt.imshow(img2)
```

### Flip on Horizontal Axis

```python
img_1 = cv2.flip(img_fix, 0)
plt.imshow(img_1)
```

### Flip on Vertical Axis

```python
img_1 = cv2.flip(img_fix, 1)
plt.imshow(img_1)
```

### Flip on Horizontal and on Vertical Axis

```python
img_1 = cv2.flip(img_fix, -1)
plt.imshow(img_1)
```

### Change the size of our canva

```python
last_img = plt.figure(figsize=(4,4))
plt1 = last_img.add_subplot(111)
plt1.imshow(img_fix)
```
